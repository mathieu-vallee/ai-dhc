```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
l = []
for j in [1,2,8]: 
    l.append(pd.read_csv('train_size%d.csv' % j))

fs = pd.concat(l)

fig, ax = plt.subplots(1, 3, figsize=(9,2))
fs['bc'].hist(ax=ax[0], bins=5, range=(0,4))
ax[0].set_xlabel('Boundary conditions ID')
ax[0].set_ylabel('Number of timeseries')
fs['eta_f_start'].hist(ax=ax[1], bins=14)
ax[1].set_xlabel('Fault start time [h]')
fs['eta_f_final'].hist(ax=ax[2], bins=20)
ax[2].set_xlabel('Fault final value [0-1]')
```
